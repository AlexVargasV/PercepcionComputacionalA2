```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image, ImageFilter
from skimage import io
from skimage import filters
```

```python
src = cv2.imread('espermatozoides.jpg',0)

plt.xticks([]), plt.yticks([])
plt.imshow(src, cmap='gray', interpolation='bicubic')
plt.title("imagen original")
plt.show()
```

```python
kernel = cv2.getStructuringElement(2,(3, 3));

dst = cv2.morphologyEx(src,0,kernel);

plt.xticks([]), plt.yticks([])
plt.imshow(dst, cmap='gray', interpolation='bicubic')
plt.title("gradiente")
plt.show()
```

```python
kernel1 = np.ones((5, 5), np.uint8)
canny = cv2.Canny(dst, 50, 150)
canny = cv2.dilate(canny, kernel1, iterations=2)
canny = cv2.erode(canny, kernel1, iterations=1)
```

```python
plt.xticks([]), plt.yticks([])
plt.imshow(canny, cmap='gray', interpolation='bicubic')
plt.title("Imagen con Clausura aplicada ")
plt.show()
```

```python
canny = cv2.erode(canny, kernel1, iterations=1)
canny = cv2.dilate(canny, kernel1, iterations=1)


plt.xticks([]), plt.yticks([])
plt.imshow(canny, cmap='gray', interpolation='bicubic')
plt.title("Imagen con Apertura aplicada ")
plt.show()
```

```python
bordercanny = cv2.Canny(canny, 50, 150)
 
plt.xticks([]), plt.yticks([])
plt.imshow(bordercanny, cmap='gray', interpolation='bicubic')
plt.title("Imagen con Canny Aplicada")
plt.show()
```

```python
img = cv2.GaussianBlur(canny,(3,3), sigmaX=0, sigmaY=0)

edge_sobel = cv2.Sobel(src=img, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=5)
plt.xticks([]), plt.yticks([])
plt.imshow(edge_sobel, cmap='gray', interpolation='bicubic')
plt.title("Imagen con Sobel Aplicada")
plt.show()
```

```python
edge_roberts = filters.roberts(canny)
plt.xticks([]), plt.yticks([])
plt.imshow(edge_roberts, cmap='gray', interpolation='bicubic')
plt.title("Imagen con Roberts Aplicada")
plt.show()
```
